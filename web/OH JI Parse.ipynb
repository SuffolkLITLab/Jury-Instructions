```python
import re
import pandas as pd
import spacy
from scipy import spatial 
nlp = spacy.load("en_core_web_lg") #this can take a minute or two to load
```

```python
from bs4 import BeautifulSoup
with open("OHJI.html") as fp3:
    soup3 = BeautifulSoup(fp3, 'html.parser')
OHcontent = soup3.find_all("div", class_ = 'ocr_carea')
```

```python
OHresults = []
for result in OHcontent: 
    result = result.text
    result = result.replace("\n", " ")
    OHresults.append(result)
OHresults = OHresults[1443:12963]
OHresults = (''.join(OHresults))
```

```python
OH_trimmed = re.sub(r"(Instructions defined)", "", OHresults)
OH_trimmed2 = re.sub(r"(/Rev. \d{2}-\d{1}-\d{2}])", "", OH_trimmed)
OH_trimmed3 = re.sub(r"(\d{1} LAW OF THE INSTRUCTIONS CV \d{3}.\d{2})", "", OH_trimmed2)
OH_trimmed4 = re.sub(r"(CV \d{3}.\d{2} OHIO JURY INSTRUCTIONS—CIVIL \d{1})", "", OH_trimmed3)
OH_trimmed5 = re.sub(r"(CV \d{3}.\d{2} OHIO JURY INSTRUCTIONS—CIVIL. \d{2})", "", OH_trimmed4)
OH_trimmed6 = re.sub(r"(CV \d{3}.\d{2} OHIO JURY INSTRUCTIONS—CIVIL \d{3})", "", OH_trimmed5)
OH_trimmed7 = re.sub(r"(\d{3} [A-Z]+ [A-Z]+ [A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed6)
OH_trimmed8 = re.sub(r"(\d{3} [A-Z]+ [A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed7)
OH_trimmed9 = re.sub(r"(\d{3} [A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed8)
OH_trimmed10 = re.sub(r"(\d{3} [A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed9)
OH_trimmed11 = re.sub(r"(\d{3} [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed10)
OH_trimmed12 = re.sub(r"(\d{3} [A-Z]+ \d{3}.\d{2})", "", OH_trimmed11)
OH_trimmed13 = re.sub(r"([A-Z]+ [A-Z]+ [A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed12)
OH_trimmed14 = re.sub(r"([A-Z]+ [A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed13)
OH_trimmed15 = re.sub(r"([A-Z]+ [A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed14)
OH_trimmed16 = re.sub(r"([A-Z]+ [A-Z]+ \d{3}.\d{2})", "", OH_trimmed15)
OHsec_results = re.split(r"(CV \d{3}.\d{2})", OH_trimmed16)
```
